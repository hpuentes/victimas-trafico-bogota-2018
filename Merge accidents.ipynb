```python
dfr = dfif.groupby(by=['GRAVEDAD','rango','ClaseNombre']).size().reset_index(name='total')
```

```python
dfHH = dfr[(dfr['GRAVEDAD'] == 'HERIDO HOSPITALIZADO')][['rango','ClaseNombre','total']]
```

```python
dfHV = dfr[(dfr['GRAVEDAD'] == 'HERIDO VALORADO')][['rango','ClaseNombre','total']]
```

```python
dfM = dfr[(dfr['GRAVEDAD'] == 'MUERTA')][['rango','ClaseNombre','total']]
```

```python
blank_data = [ ('<= 4 años',0,0,0,0,0,0),
             ('5-15 años',0,0,0,0,0,0),
             ('16-23 años',0,0,0,0,0,0),
             ('24-30 años',0,0,0,0,0,0),
             ('31-40 años',0,0,0,0,0,0),
             ('41-50 años',0,0,0,0,0,0),
             ('51-60 años',0,0,0,0,0,0),
             ('61-70 años',0,0,0,0,0,0),
             ('71-80 años',0,0,0,0,0,0),
             ('>= 81 años',0,0,0,0,0,0)]
```

```python
column_names = ['rango','Choque','Atropello','Caida de ocupante','Volcamiento','Incendio','Otro']
```

```python
dfHHr = pd.DataFrame(blank_data,columns=column_names)
dfHVr = pd.DataFrame(blank_data,columns=column_names)
dfMr = pd.DataFrame(blank_data,columns=column_names)
```

```python
def load_Data(data,target):
    for index, row in data.iterrows():
        itrango = row['rango']
        for index2, row2 in target.iterrows():
            if row2['rango'] == itrango:
                target.at[index2,row['ClaseNombre']] = row['total']
```
